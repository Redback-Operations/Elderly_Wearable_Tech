```python
import pandas as pd
import os

directory = r"C:\Users\asus\Desktop\Elderly_Wearable_Tech\Data Analytics\Most common Disease feature\dataset"

# List to hold data from each file
all_data = []

# Iterate through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.xlsx'):
        file_path = os.path.join(directory, filename)
        df = pd.read_excel(file_path)
        if not all_data or (all_data and list(df.columns) == list(all_data[0].columns)):
            all_data.append(df)
        else:
            print(f"Headers do not match in file: {filename}")
            # Handle the case when headers do not match

# Combine all DataFrames into one
combined_df = pd.concat(all_data, ignore_index=True)

# Save the combined DataFrame to a new Excel file
output_file = os.path.join(directory, '2014-2019.xlsx')
combined_df.to_excel(output_file, index=False)
print("Combined Excel file created.")

# specific path for the xlsx file
xlsx_path = r"C:\Users\asus\Desktop\Elderly_Wearable_Tech\Data Analytics\Most common Disease feature\dataset\2014-2019.xlsx"
# Specify the output CSV file path
csv_path = r"C:\Users\asus\Desktop\Elderly_Wearable_Tech\Data Analytics\Most common Disease feature\dataset\2014-2019.csv"

# Read the Excel file into a pandas DataFrame
df = pd.read_excel(xlsx_path)
# Save the DataFrame to a CSV file
df.to_csv(csv_path, index=False)

print(f"File has been converted to CSV and saved at {csv_path}")
```

```python
import pandas as pd
import numpy as np

# Loading the dataset
file_path = r'C:\Users\asus\Desktop\Elderly_Wearable_Tech\Data Analytics\Most common Disease feature\dataset\2014-2019.csv'
data = pd.read_csv(file_path)

# Renaming the 'ISO3' column to 'Country'
data.rename(columns={'ISO3': 'Country'}, inplace=True)

# Replacing values in the 'Sex' column
data['Sex'] = data['Sex'].replace({'MLE': 'M', 'FMLE': 'F'})


# Allocating the random ages within the specified ranges in the 'Age Group' column
def allocate_random_age(age_group):
    if age_group.startswith("YEARS"):
        age_range = age_group[5:].split("-")
        if len(age_range) == 2:
            return np.random.randint(int(age_range[0]), int(age_range[1]) + 1)
        else:
            return np.nan
    else:
        return np.nan  
        
data['Age Group'] = data['Age Group'].apply(allocate_random_age)

# Displaying the head to verify changes
data.head()
```
