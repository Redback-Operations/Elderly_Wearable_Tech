```python
import os
import pandas as pd
```

```python
FOLDER_PATH = "Dataset"
OUTPUT_FOLDER_PATH = "CSV"

txt_files = [f for f in os.listdir(FOLDER_PATH) if f.endswith('.txt')]

for txt_file in txt_files:
    INPUT_FILEA_PATH = os.path.join(FOLDER_PATH, txt_file)
    OUTPUT_FILE_PATH = os.path.join(OUTPUT_FOLDER_PATH,
                                    txt_file.replace('.txt', '.csv'))

    df = pd.read_csv(INPUT_FILEA_PATH, delimiter=' ')
    df.to_csv(OUTPUT_FILE_PATH, index=False)
    print(f"Converted {txt_file} to {OUTPUT_FILE_PATH}")

print("All files converted successfully!")
```

```python
# Constants for folder and output paths
FOLDER_PATH = 'CSV/PD'
OUTPUT_PATH = 'CSV/PD.csv'

PD_subjects = pd.DataFrame()

for filename in os.listdir(FOLDER_PATH):
    if filename.endswith('.csv'):
        file_path = os.path.join(FOLDER_PATH, filename)
        data = pd.read_csv(file_path)
        PD_subjects = pd.concat([PD_subjects, data], ignore_index=True)

PD_subjects.to_csv(OUTPUT_PATH, index=False)
print(f"All CSV files merged successfully into {OUTPUT_PATH}")
```
