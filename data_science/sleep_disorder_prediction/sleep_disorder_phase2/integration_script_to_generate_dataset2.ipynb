# Integartion of feature extraction script to generate phase2 dataset.


Phase 2's Dataset 2 is generated by extracting sleep metrics from individual subjects' CSV files using the provided code. The code calculates various sleep statistics, including total sleep duration, sleep latency, time spent in different sleep stages, and the Arousal Index (AI), compiling these metrics for each subject. These results are consolidated into a single dataset for further analysis and predictive modeling.

```python
import os
import pandas as pd

# Step 1: Load the Excel file with multiple sheets
# uplaod the dataset from which the features for dataset2 should be extracted.

file_path = "data/SleepStagingArterialFeatures.xlsx"  # Update with your Excel file path
output_folder = "data/subjects"  # Folder to save CSV files

# Create output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Get the sheet names from the Excel file
excel_file = pd.ExcelFile(file_path)
sheet_names = excel_file.sheet_names  # Get a list of all sheet names

# Step 2: Extract each sheet and save as an individual CSV
for sheet_name in sheet_names:
    data = pd.read_excel(file_path, sheet_name=sheet_name)  # Read each sheet from Excel
    csv_filename = f"{output_folder}/{sheet_name}.csv"
    data.to_csv(csv_filename, index=False)  # Save as a CSV file

    print(f"Saved {sheet_name}.csv to {output_folder}")
```

```python
import pandas as pd

# Load the dataset
def calculate_sleep_metrics(csv_filename):
    data = pd.read_csv(csv_filename)
    data

    # Find the first occurrence of Class 6
    first_occurrence_index = data[data['Class'] == 6].index.min()
    temp_index = None  # Variable to store the index where 3 is found initially

    # Iterate through 'Class' to find the first occurrence of values in [1,2,3,4] after Class 6
    for index, value in enumerate(data['Class']):
        if value in [1, 2, 3, 4] and temp_index is None:
            temp_index = index
            break
        elif value == 6:
            temp_index = None

    # Calculate the time difference between Class 6 and the first occurrence of [1, 2, 3, 4]
    segment_difference = temp_index - first_occurrence_index
    time_difference_seconds = segment_difference * 30

    # Convert time difference to hours, minutes, and seconds
    hours, remainder = divmod(time_difference_seconds, 3600)
    minutes, seconds = divmod(remainder, 60)

    # Output the time difference
    print("Time difference:", int(hours), "hours,", int(minutes), "minutes and", int(seconds), "seconds.")

    # Initialize total time variables
    total_time = 0
    index = 0  # Initialize index

    # Loop through the 'Class' column and calculate total time for each class in [1,2,3,5]
    while index < len(data['Class']):
        start_index = None
        end_index = None

        if data['Class'][index] in [1, 2, 3, 4, 5]:
            start_index = index
            while index < len(data['Class']) and data['Class'][index] in [1, 2, 3, 5]:
                index += 1
            end_index = index - 1
            if start_index is not None and end_index >= 0:
                segment_difference = end_index - start_index + 1
                total_time += segment_difference * 30  # Each segment is 30 seconds
        index += 1
        

    # Convert total time to hours, minutes, and seconds
    hours, remainder = divmod(total_time, 3600)
    minutes, seconds = divmod(remainder, 60)
    sleep_duration_minutes = total_time//60

    # Output total sleep time
    sleep_duration = (int(hours), "hours,", int(minutes), "minutes and", int(seconds), "seconds.")
    print(sleep_duration)
    print("Total sleep time:", int(hours), "hours,", int(minutes), "minutes and", int(seconds), "seconds.")

    # Calculate time difference between first occurrence of Class 5 and [1,2,3,4]
    first_occurrence_index = data[data['Class'].isin([1, 2, 3, 4])].index.min()
    timestamp2_index = first_occurrence_index

    first_occurrence_index = data[data['Class'] == 5].index.min()
    timestamp1_index = first_occurrence_index

    segment_difference = timestamp1_index - timestamp2_index
    time_difference_seconds = segment_difference * 30
    sleep_difference_minutes = time_difference_seconds//60

    # Convert time difference to hours, minutes, and seconds
    hours, remainder = divmod(time_difference_seconds, 3600)
    minutes, seconds = divmod(remainder, 60)

    # Output the time difference
    sleep_difference = (int(hours), "hours,", int(minutes), "minutes and", int(seconds), "seconds.")
    print(sleep_difference)
    print("Time difference:", int(hours), "hours,", int(minutes), "minutes and", int(seconds), "seconds.")

    # Initialize total times for each class
    total_time_1 = 0  # NoREM 1
    total_time_2 = 0  # NoREM 2
    total_time_3 = 0  # NoREM 3
    total_time_4 = 0  # NoREM 4
    total_time_rem = 0  # REM (Class 5)

    index = 0  # Initialize index
    total_segments = len(data['Class'])

    # Loop through the 'Class' column to calculate total time for each stage
    while index < total_segments:
        start_index = None
        end_index = None

        # NoREM 1 (Class 1)
        if data['Class'][index] == 1:
            start_index = index
            while index < total_segments and data['Class'][index] == 1:
                index += 1
            end_index = index - 1
            if start_index is not None and end_index >= 0:
                segment_difference = end_index - start_index + 1
                total_time_1 += segment_difference * 30

        # NoREM 2 (Class 2)
        elif data['Class'][index] == 2:
            start_index = index
            while index < total_segments and data['Class'][index] == 2:
                index += 1
            end_index = index - 1
            if start_index is not None and end_index >= 0:
                segment_difference = end_index - start_index + 1
                total_time_2 += segment_difference * 30

        # NoREM 3 (Class 3)
        elif data['Class'][index] == 3:
            start_index = index
            while index < total_segments and data['Class'][index] == 3:
                index += 1
            end_index = index - 1
            if start_index is not None and end_index >= 0:
                segment_difference = end_index - start_index + 1
                total_time_3 += segment_difference * 30

        # NoREM 4 (Class 4)
        elif data['Class'][index] == 4:
            start_index = index
            while index < total_segments and data['Class'][index] == 4:
                index += 1
            end_index = index - 1
            if start_index is not None and end_index >= 0:
                segment_difference = end_index - start_index + 1
                total_time_4 += segment_difference * 30

        # REM (Class 5)
        elif data['Class'][index] == 5:
            start_index = index
            while index < total_segments and data['Class'][index] == 5:
                index += 1
            end_index = index - 1
            if start_index is not None and end_index >= 0:
                segment_difference = end_index - start_index + 1
                total_time_rem += segment_difference * 30

        index += 1

    # Calculate total sleep time across all classes
    total_time = total_time_1 + total_time_2 + total_time_3 + total_time_4 + total_time_rem

    # Convert total times to hours, minutes, and seconds
    def convert_time(seconds):
        hours, remainder = divmod(seconds, 3600)
        minutes, seconds = divmod(remainder, 60)
        return int(hours), int(minutes), int(seconds)

    # Print time for each sleep stage
    for stage, time in zip(
        ["NoREM 1", "NoREM 2", "NoREM 3", "NoREM 4", "REM"],
        [total_time_1, total_time_2, total_time_3, total_time_4, total_time_rem]
    ):
        hours, minutes, seconds = convert_time(time)
        print(f"Time difference {stage}: {hours} hours, {minutes} minutes and {seconds} seconds.")

    # Calculate percentage for each stage
    def calculate_percentage(total_time, class_time):
        return (class_time / total_time) * 100 if total_time != 0 else 0

    percent_1 = calculate_percentage(total_time, total_time_1)
    percent_2 = calculate_percentage(total_time, total_time_2)
    percent_3 = calculate_percentage(total_time, total_time_3)
    percent_4 = calculate_percentage(total_time, total_time_4)
    percent_rem = calculate_percentage(total_time, total_time_rem)

    # Output the percentage for each stage
    print(f"Percentage of NoREM 1: {percent_1:.2f}%")
    print(f"Percentage of NoREM 2: {percent_2:.2f}%")
    print(f"Percentage of NoREM 3: {percent_3:.2f}%")
    print(f"Percentage of NoREM 4: {percent_4:.2f}%")
    print(f"Percentage of REM: {percent_rem:.2f}%")

    # Arousal index calculation
    arousal_count = 0
    total_sleep_time = 0
    index = 0

    # Loop through the dataset
    while index < len(data['Class']):
        if data['Class'][index] in [2, 3, 4, 5]:
            total_sleep_time += 30
            if index < len(data['Class']) - 1 and data['Class'][index + 1] in [1, 6]:
                arousal_count += 1
        index += 1

    # Convert total sleep time to hours
    total_sleep_time_hours = total_sleep_time / 3600

    # Calculate and output the Arousal Index (arousals per hour of sleep)
    arousal_index = arousal_count / total_sleep_time_hours if total_sleep_time_hours > 0 else 0
    print("Arousal Index (AI):", arousal_index, "arousals per hour")
    return {
        "Subject": csv_filename.split('/')[-1].split('.')[0],  # Extract subject ID from filename
        "Sleep Duration (hours)": sleep_duration_minutes,
        "Sleep Latency (minutes)": sleep_difference_minutes,
        "REM1% (minutes)": percent_1,
        "REM2% (minutes)": percent_2,
        "REM3% (minutes)": percent_3,
        "REM %":percent_rem,
        "Arousal Index": arousal_index
    }
```

```python
# Step 4: Run the calculation on each subject CSV and combine results
all_subjects_data = []
for sheet_name in sheet_names:
    csv_filename = f"{output_folder}/{sheet_name}.csv"
    subject_metrics = calculate_sleep_metrics(csv_filename)
    all_subjects_data.append(subject_metrics)

# Step 5: Convert the results into a DataFrame and save as a new CSV
dataset2 = pd.DataFrame(all_subjects_data)
dataset2.to_csv("data/phase2_dataset/dataset2.csv", index=False)

print("All subjects processed and dataset2 created.")
```
